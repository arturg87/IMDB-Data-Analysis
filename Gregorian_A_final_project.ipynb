# Background

Is it possible to predict the success of a movie based on a certain set of data points? That is the purpose of this study. 

The goal of this study is to predict the success of a movie by answering the following questions: 
1. Does it matter who the director is for a movie to be successful? 
2. Does it matter who the actor or actress is for a movie to be successful? 
3. What genre of movies are most successful? 
4. How does the movie rating (PG, PG-13, R, etc) affect the success of a movie? 
5. Do facebook likes determine movie success? 
6. Does a successful movie have and Oscar wins or Nominations? 
7. Is the IMDB score a good indicator of movie success? 
    1. What variables are highly correlated with the IMDB score? 
    2. Look into the number of votes used to create the score.

## Data Description
The data set in use is collected from the Internet Movie DataBase (IMDB) website. It comprises of the following data:
[Data source](https://data.world/popculture/imdb-5000-movie-dataset)

The original dataset was collected in two parts: 
1. The names of the movies were first collected from the-numbers.com. 
2. Second, the names collected in the previous step were used to search IMDB for scraping.
This method worked fine, however, upon closer inspection of the data, I found it to be a bit misleading. When searching IMDB with the movie title from the-numbers.com, it didn't always result in the feature film on the IMDB website. Sometimes, IMDB returned a TV show or video game instead and this data was scraped. This was due to the slightly different movie title across the pages.

To remedy this, I downloaded the code provided by the original author of the dataset ([source](https://github.com/sundeepblue/movie_rating_prediction/blob/master/movie/spiders/imdb_url_spider.py))
to make the following changes:
1. I first searched IMDB for all feature films and sorted the results by Gross profit. 
2. Next the scraped the top 5000 search results to collect the Movie urls and titles
3. Lastly, I scarped the individual movie page in IMDB from the urls collected in step 2.
4. Additionally, I added data to show how many Oscars it had won. If the movie had no wins, I collected data on how many Oscars it was nominated for.

## Does the director or cast affect the IMDB score? 

```python
# This function will create a dataframe that includes number of occurances and average IMDB score of the person passed into the function
# This function takes has two inputs: ds is the dataset and type_name is either director or actors.
def name_avg_IMDB_score(ds, type_name, type_calc="avg"):
   
    # Iterate through all director names
    if type_name == "director":
        avg = []
        name = []
        occurrences = []
        values = ds.value_counts().keys().tolist()
        counts = ds.value_counts().tolist()
        for i in range(len(values)):
            # store number of occurance name shows up in dataset
            occurrences = occurrences + [counts[i]]
            
            if type_calc == "avg":
                
                # store average imdb_score (averaged across all occurances of the passed in name)
                avg = avg + [imdb[ds==values[i]].imdb_score.mean()]
            elif type_calc == "sum":
                avg = avg + [imdb[ds==values[i]].imdb_score.sum()]
            # Store name
            name = name + [values[i]]
    # Iterate through all actors names. This will need to be its own Dataset. 
    # The original dataset has the actors split into three columns. This function will parse all actors names, so we will make a
    # dataframe with all three columns of name and thier IMDB score.
    elif type_name == "cast":
        avg = []
        name = []
        occurrences = []
        # Get name values from value_counts
        values = ds.value_counts().keys().tolist()
        # Get the number of counts for name value from value_counts
        counts = ds.value_counts().tolist()
        # Make sure all elements are integar
        counts = [ int(x) for x in counts]
        
        # populate data
        for i in range(len(values)):
            occurrences = occurrences + [counts[i]]
            avg = avg +[merged[ds==values[i]].imdb_score.mean()]
            name = name + [values[i]]
            
    # Zip name, occurance and averages together
    d = zip(name, occurrences, avg)
    
    # Create dataframe from zipped data
    df = pd.DataFrame(d, columns=['Name', 'Occurrences', 'Avg IMDB Score']).set_index('Name')
    
    # Return dataframe
    return df
```

## How does the genre relate to a movie's success?

```python
# This function counts the number of genres filtered by a specific imdb score that is passed into the function=
# n is a single numbe =r to look for
# c is the column name we want to search in.
# r1 is the first range value
# r2 is the second range value
def genre_filter(c, n=0, r1=0, r2=0, all_genres=0):
    # List of genres we will be searching for in dataset
    genre_list = ['Drama', 'Comedy', 'Thriller', 'Action' ,'Romance' ,'Adventure','Crime','Family','Fantasy','Sci-Fi','Mystery','Horror','Animation','Biography','Sport','Music','History','War','Musical','Western','Film-Noir']
    
    genre = []
    score = []

    # Iterate through list of genres and search for the occurance of that genre
    # Additionally filter rows that are below the passed in score
    for i in range(len(genre_list)):
        # if a range is specified, filter by the specified range.
        if r1:
            filt2 = (imdb['genres'].str.contains(genre_list[i], regex=False)) & (imdb[c] >= r1) & (imdb[c] <= r2)
        elif all_genres:
            filt2 = (imdb['genres'].str.contains(genre_list[i], regex=False))
        else: 
            filt2 = (imdb['genres'].str.contains(genre_list[i], regex=False)) & (imdb[c] >= n) 
        df2 = imdb[filt2].sort_values('imdb_score', ascending=False)
        # get the count of the number of occurances of genres
        count = df2['genres'].count()
        #if count is 0, make it a NaN so we can drop it later
        if count == 0:
            genre = genre + [genre_list[i]]
            score = score + [np.NaN]
        else:
        # else add the count to the appropriate place in the new dataframe 
            genre = genre + [genre_list[i]]
            score = score + [count]
    d = zip(genre, score)
    df = pd.DataFrame(d, columns=['Genre', 'Count']).set_index('Genre')
    # drop any null values
    df = df.dropna()
    # sort values in the data frame
    df = df.sort_values('Count', ascending=True)
    # return the data frame
    return df
```

```python
ax = genre_filter("profit", all_genres='y').plot(kind='barh', figsize=(20,10), legend=False)
ax.set_title("Count of all genres.", fontsize=20)
ax.set_xlabel("Number of Films", fontsize=15)
ax.set_ylabel("Genres", fontsize=15)
ax.tick_params(labelsize=15)

for p in ax.patches:
    ax.annotate(str(p.get_width()), (p.get_width()+1, p.get_y()+.1), fontsize=12)
    
```

Since we have not accomplished ETL up to this stage, it is necessary to clean the data for modeling purposes. The data, as it is, is already fairly clean due to the additional steps taken to re-scrape the data at the beginning of the Mid-Term project. However, there is still 767 rows missing data in the budget column. I decided that I didn't want to lost that data by dropping rows, so I filled all null data in the bugdet, profit, and gross columns with 0 and dropped all other null rows. 

Additionally, we need to transform some of the categorical values to numerical values for those columns to be considered as a predictor. However, due to the number of variation in some of the columns, we will not need the following columns and will be dropping those as well. 
* movie_title
* director_name
* actor_1_name

* actor_3_name

* country

```python
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# We need to convert certain text to int for it to be considered as a predictor. 
# Titles and Director\Actor names will be dropped as well as plot_keywords.

data = imdb

# First fill budget, gross, and profit with 0
# this is the bulk of the null data and i did not want to lose it.
data = data.fillna({'budget': 0, 'gross': 0, 'profit': 0})

# Second, drop the rest of the null data
data = data.dropna()


# Drop Movie Title, Director Name, Actor 1-3 name, Plot Keywords, Genres, Langauge, Country and Movie Link
data = data.drop(['movie_title', 'director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'plot_keywords', 'language', 'country', 'movie_imdb_link'], axis=1)

# map categorical value to numerical values
data['color'] = data['color'].map({'Black and White': 0, 'Color': 1})

# We only care about content ratings of G, PG, PG-13, R. Dropping the ones we don't care about will only drop 109 entries. 
to_keep = ['PG-13', 'PG', 'G', 'R']
data = data.loc[data.content_rating.isin(to_keep)]

# Map Content Rating categorical values to numerical values
data['content_rating'] = data['content_rating'].map({'G': 0, 'PG': 1, 'PG-13': 2, 'R': 3})
```

```python
data.head()
```

```python
# Convert the string to list 
# https://stackoverflow.com/questions/50278300/convert-a-columns-of-string-to-list-in-pandas
data['genres'] = data['genres'].str.split(',')

# One hot encode columns with a list of strings as values
# https://stackoverflow.com/questions/47786822/how-do-you-one-hot-encode-columns-with-a-list-of-strings-as-values
df1 = data['genres'].str.join('|').str.get_dummies()

# Join dummies to orignal data
data = data.join(df1)

# drop genres column
data = data.drop('genres', axis=1)

# Remove dash from Sci-Fi for later
data = data.rename({'Sci-Fi': 'SciFi'}, axis=1)
data.head()
```

```python
from sklearn.metrics.pairwise import pairwise_distances_argmin
from scipy.spatial.distance import cdist 
from sklearn import metrics
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm

from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import ward_tree
from scipy.cluster.hierarchy import dendrogram, linkage, ward

from sklearn import tree

#The Elbow method

#https://www.packtpub.com/big-data-and-business-intelligence/mastering-machine-learning-scikit-learn#
#http://scikit-learn.org/stable/modules/clustering.html

K = range(1, 10) 
meandistortions = [] 
for k in K: 
    kmeans = KMeans(n_clusters=k) 
    kmeans.fit(x) 
    meandistortions.append(sum(np.min(cdist(x, kmeans.cluster_centers_, 'euclidean'), axis=1)) / x.shape[0]) 

plt.plot(K, meandistortions, 'bx-') 
plt.xlabel('k') 
plt.ylabel('Average distortion') 
plt.title('Selecting k with the Elbow Method') 
plt.show()
```

```python
plt.figure(figsize=(16,8))

plt.title('Hierarchical Clustering Dendrogram (truncated)')
plt.xlabel('sample index or (cluster size)')
plt.ylabel('distance')

linkage_matrix = ward(df_norm)
dendrogram(linkage_matrix, 
           truncate_mode='lastp',  # show only the last p merged clusters
           p=2,  # show only the last p merged clusters
           #show_leaf_counts=False,  # otherwise numbers in brackets are counts
           leaf_rotation=90.,
           leaf_font_size=12.,
           show_contracted=True,  # to get a distribution impression in truncated branches
           orientation="top")
plt.tight_layout() # fixes margins
```
